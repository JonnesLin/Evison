```python
from display import Display
from torchvision import models
```

### Load Model

```python
resnet18 = models.resnet18(pretrained=True)
```

### Parameters for Display

##### The fisrt parameters is model
##### The second parameters is module name
##### The third parameters is the index of layer in the module

```python
print(resnet18)
```

### The first thing we need to do is to get the module name and the index of that
#### Below:
#####     conv1, bn1, relu, maxpool, layer1, layer2, layer3, layer4, avgpool, fc are module name
#####     the index of layer is the number in the module
#### Example:
#####    If I want get the first block of the layer2, we can set the module name is layer2 and the layer index is 0



```python
module_name = 'layer2'
layer_index = 0

shower = Display(resnet18, module_name, layer_index)
```

```python
# Load the image you want to show
from PIL import Image

image = Image.open('dogs_1280p_0.jpeg').resize((224, 224))
image
```

```python
# save the result
shower.save(image)
```
