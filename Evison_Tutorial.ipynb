```python
from Evison import Display, show_network
from torchvision import models
```

```python
# create efficientnet model 
network = models.efficientnet_b0(pretrained=True)

# show which layer we can visualize
show_network(network)
```

```python
# create visualization object
visualized_layer = 'features.7.0'
display = Display(network, visualized_layer, img_size=(224, 224))
```

```python
# Load the image you want to show
from PIL import Image

image = Image.open('Dog_and_cat.jpeg').resize((224, 224))
image
```

```python
# save the result
display.save(image)
```
